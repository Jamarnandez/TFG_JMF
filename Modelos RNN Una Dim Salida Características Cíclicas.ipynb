# Definición de funciones

```python
################################# DEFINICIÓN DE FUNCIONES #################################

#########################################################
## Lectura de datos desde csv del repositorio de datos ##
#########################################################
def lee_datos(files_var):
  dataset = pd.read_csv(files[0], sep=';', dtype = {'PROVINCIA': int, 'MUNICIPIO': int, 'ESTACION': int, 'MAGNITUD': int
                                                    , 'PUNTO_MUESTREO': str, 'ANO': int, 'MES': int, 'DIA': int
                                                    , 'H01': float64, 'V01': str, 'H02': float64, 'V02': str, 'H03': float64, 'V03': str
                                                    , 'H04': float64, 'V05': str, 'H05': float64, 'V05': str, 'H06': float64, 'V06': str
                                                    , 'H07': float64, 'V07': str, 'H08': float64, 'V08': str, 'H09': float64, 'V09': str
                                                    , 'H10': float64, 'V10': str, 'H11': float64, 'V11': str, 'H12': float64, 'V12': str
                                                    , 'H13': float64, 'V13': str, 'H14': float64, 'V14': str, 'V15': float64, 'V15': str
                                                    , 'H16': float64, 'V16': str, 'H17': float64, 'V17': str, 'H18': float64, 'V18': str
                                                    , 'H19': float64, 'V19': str, 'H20': float64, 'V20': str, 'H21': float64, 'V21': str
                                                    , 'H22': float64, 'V22': str, 'H23': float64, 'V23': str, 'H24': float64, 'V24': str})
  df_concat = dataset
  for i in range(1, len(files)):
    dataset = pd.read_csv(files[i], sep=';', dtype = {'PROVINCIA': int, 'MUNICIPIO': int, 'ESTACION': int, 'MAGNITUD': int
                                                    , 'PUNTO_MUESTREO': str, 'ANO': int, 'MES': int, 'DIA': int
                                                    , 'H01': float64, 'V01': str, 'H02': float64, 'V02': str, 'H03': float64, 'V03': str
                                                    , 'H04': float64, 'V05': str, 'H05': float64, 'V05': str, 'H06': float64, 'V06': str
                                                    , 'H07': float64, 'V07': str, 'H08': float64, 'V08': str, 'H09': float64, 'V09': str
                                                    , 'H10': float64, 'V10': str, 'H11': float64, 'V11': str, 'H12': float64, 'V12': str
                                                    , 'H13': float64, 'V13': str, 'H14': float64, 'V14': str, 'V15': float64, 'V15': str
                                                    , 'H16': float64, 'V16': str, 'H17': float64, 'V17': str, 'H18': float64, 'V18': str
                                                    , 'H19': float64, 'V19': str, 'H20': float64, 'V20': str, 'H21': float64, 'V21': str
                                                    , 'H22': float64, 'V22': str, 'H23': float64, 'V23': str, 'H24': float64, 'V24': str})
    df_concat = pd.concat([df_concat, dataset]).reset_index(drop=True).drop_duplicates()

  return df_concat




#######################################################################
## La función filtra_estaciones(df) elimina todas las estaciones     ##
## que no van a ser objeto de estudio en este trabajo. en este       ##
## caso, se han seleccionado las estaciones:                         ##
##      -> 28079004 - Plaza de España                                ##
##      -> 28079008 - Escuelas Aguirre                               ##
##      -> 28079048 - Paseo Castellana                               ##
#######################################################################
def filtra_estaciones(df):
  g = 0
  while len(df) > 3:
    if df[g] != est_pza_espania and df[g] != est_escuelas_aguirre and df[g] != est_paseo_castellana:
      df.remove(df[g])
    else:
      g+=1
  return df

##################################################################
## La función time_serie(df) convierte el dataframe de entrada  ##
## en una serie temporal, creando la columna DATETIME calculada ##
## a partir de las columnas FECHA y las Columnas que definimos  ##
## previamente en el diccionario creado en el paso anterior     ##
##################################################################
def time_serie(dataset):
  for i in range(1,25):
    dataset_filt = dataset[(dataset[dicV[i]] == 'V')]

    dataset_AUX = dataset_filt[["FECHA", "ESTACION", "MAGNITUD", dicH[i]]]
    if i == 24:
      y = '00'
    else:
      y = str(i)


    dataset_AUX["DATETIME"] = dataset_AUX.apply(lambda x: str(x["FECHA"])[:10] + ' ' + y +':00:00', axis=1)

    dataset_AUX["DATETIME"] = pd.to_datetime(dataset_AUX.loc[:,"DATETIME"], format='%y-%m-%d %H:%M:%S', yearfirst=True, infer_datetime_format=True)

    dataset_AUX.drop('FECHA', axis=1, inplace=True)
    column = str(dicH[i])
    dataset_AUX.rename({column:'MUESTRA'}, axis=1, inplace=True)

    if i == 1:
      dfFinal = dataset_AUX
    else:
      dfFinal = pd.concat([dfFinal, dataset_AUX]).reset_index(drop=True)

  return dfFinal

####################################################################
## La función time_serie_varias_estaciones(dataframe, estaciones) ##
## se crea como una función auxiliar para llamar a la función     ##
## time_serie y así formatear todas las estaciones recogidas en   ##
## el orígen de datos                                             ##
####################################################################
def time_serie_varias_estaciones(dataframe, estaciones):
  i = 1
  for estacion in estaciones:
    df = dataframe[(dataframe['ESTACION'] == estacion)]

    df_time_serie = time_serie(df)

    if i == 1:
      dfFinal = df_time_serie
    else:
      dfFinal = pd.concat([dfFinal, df_time_serie]).reset_index(drop=True)

    i += 1

  return dfFinal


########################################################
## La función crea_calendario(list) se utiliza para   ##
## crear la línea temporal del año y se le añade la   ##
## información de los distintos sensores.             ##
########################################################
def crea_calendario(estaciones, df_datos):
  i = 1
  date_df = pd.date_range(start=start_date, end=end_date, freq="H").astype(str).to_frame(index=False, name="DATETIME")
  for estacion in estaciones:

    data_df = df_datos[(df_datos['ESTACION'] == estacion)]
    data_df = data_df[["DATETIME", "MUESTRA"]]
    data_df = data_df.rename(columns={'MUESTRA':f'''ESTACION_{str(estacion)}'''})
    data_df["DATETIME"] = pd.to_datetime(data_df["DATETIME"], yearfirst=True)
    date_df["DATETIME"] = pd.to_datetime(date_df["DATETIME"], yearfirst=True)

    if i == 1:
      merge_df = date_df.merge(data_df, on=["DATETIME"], how="left")
    else:
      merge_df = merge_df.merge(data_df, on=["DATETIME"], how="left")

    i += 1

  return merge_df



##########################################################
## Definición de la función que usaremos para pintar la ##
## gráfica y mostrar los datos de manera visual         ##
##########################################################
def plot_dataset(df, title):
    data = []
    value1 = go.Scatter(
        x=df.index,
        y=df.value1,
        mode="lines",
        name="ESTACION_4",
        marker=dict(),
        text=df.index,
        line=dict(color="rgba(200,200,100, 1)"),
    )
    data.append(value1)

    value2 = go.Scatter(
        x=df.index,
        y=df.value2,
        mode="lines",
        name="ESTACION_8",
        marker=dict(),
        text=df.index,
        line=dict(color="rgba(100,100,100, 1)"),
    )
    data.append(value2)

    value3 = go.Scatter(
        x=df.index,
        y=df.value3,
        mode="lines",
        name="ESTACION_48",
        marker=dict(),
        text=df.index,
        line=dict(color="rgba(100,300,200, 1)"),
    )
    data.append(value3)

    layout = dict(
        title=title,
        xaxis=dict(title="Date", ticklen=5, zeroline=False),
        yaxis=dict(title="Polutant", ticklen=5, zeroline=False),
    )

    fig = dict(data=data, layout=layout)
    iplot(fig)


##########################################################
## Definición de la función que usaremos para realizar  ##
## el cálculo de las métricas del modelo                ##
##########################################################
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def calculate_metrics(df):
    result_metrics = {'mae' : mean_absolute_error(df.value, df.prediction),
                      'rmse' : mean_squared_error(df.value, df.prediction) ** 0.5,
                      'r2' : r2_score(df.value, df.prediction)}

    print("Mean Absolute Error:       ", result_metrics["mae"])
    print("Root Mean Squared Error:   ", result_metrics["rmse"])
    print("R^2 Score:                 ", result_metrics["r2"])
    return result_metrics
```

# Generación ventana deslizante con uno de los sensores

```python
df_lag = interpolated_stations["ESTACION_8"].to_frame()
input_dim = rows = 96

df_generated = df_lag.copy()
```

```python
def generate_time_lags(df, n_lags):
    df_n = df.copy()
    for n in range(1, n_lags + 1):
        df_n[f"lag{n}"] = df_n["ESTACION_8"].shift(n)
    df_n = df_n.iloc[n_lags:]
    return df_n



df_generated = generate_time_lags(df_lag, input_dim)
df_generated = df_generated.rename(columns={'ESTACION_8':'value'})
#df_generated
```

```python
df_generated = (
                df_generated
                .assign(hour = df_generated.index.hour)
                .assign(day = df_generated.index.day)
                .assign(month = df_generated.index.month)
                .assign(day_of_week = df_generated.index.dayofweek)
                .assign(week_of_year = df_generated.index.week)
              )
```

```python
import numpy as np

def generate_cyclical_features(df, col_name, period, start_num=0):
    kwargs = {
        f'sin_{col_name}' : lambda x: np.sin(2*np.pi*(df[col_name]-start_num)/period),
        f'cos_{col_name}' : lambda x: np.cos(2*np.pi*(df[col_name]-start_num)/period)
             }
    return df.assign(**kwargs).drop(columns=[col_name])

df_generated = generate_cyclical_features(df_generated, 'hour', 23, 0)
df_generated = generate_cyclical_features(df_generated, 'day', 31, 0)
df_generated = generate_cyclical_features(df_generated, 'month', 12, 1)
df_generated = generate_cyclical_features(df_generated, 'day_of_week', 7, 0)
df_generated = generate_cyclical_features(df_generated, 'week_of_year', 52, 0)
df_generated = generate_cyclical_features(df_generated, 'year', 2023, 2018)

df_generated
```

# División del conjunto de datos

Se divide el conjunto de datos en 3 conjuntos: entrenamiento, validación y prueba.

```python
from sklearn.model_selection import train_test_split

def feature_label_split(df, target_col):
    y = df[[target_col]]
    X = df.drop(columns=[target_col])
    return X, y

def train_val_test_split(df, target_col, test_ratio):
    val_ratio = test_ratio / (1 - test_ratio)
    X, y = feature_label_split(df, target_col)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_ratio, shuffle=False)
    X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=val_ratio, shuffle=False)
    return X_train, X_val, X_test, y_train, y_val, y_test

X_train, X_val, X_test, y_train, y_val, y_test = train_val_test_split(df_generated, 'value', 0.1)
```

# Creación de la redes neuronales


---



```python
class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, layer_dim, output_dim, dropout_prob):
        super(LSTMModel, self).__init__()

        # Definir el número de capas y los nodos de cada capa
        self.hidden_dim = hidden_dim
        self.layer_dim = layer_dim

        # Capas LSTM
        self.lstm = nn.LSTM(
            input_dim, hidden_dim, layer_dim, batch_first=True, dropout=dropout_prob
        )

        # Capa totalmente conectada
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        # Inicializando el estado oculto para la primera entrada con ceros
        h0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).requires_grad_()

        # Inicializando el estado de la celda para la primera entrada con ceros
        c0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).requires_grad_()

        # Necesitamos separarnos ya que estamos haciendo retropropagación truncada en el tiempo (BPTT)
        # Si no lo hacemos, retropropagaremos todo el camino hasta el principio incluso después de pasar por otro lote
        # Propagación hacia adelante pasando la entrada, el estado oculto y el estado de la célula en el modelo
        out, (hn, cn) = self.lstm(x, (h0.detach(), c0.detach()))

        # Remodelar las salidas en forma de (batch_size, seq_length, hidden_size)
        # para que quepa en la capa totalmente conectada
        out = out[:, -1, :]

        # Convertir el estado final a nuestra forma de salida deseada (batch_size, output_dim)
        out = self.fc(out)

        return out
```

```python
class GRUModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, layer_dim, output_dim, dropout_prob):
        super(GRUModel, self).__init__()

        # Definir el número de capas y los nodos de cada capa
        self.layer_dim = layer_dim
        self.hidden_dim = hidden_dim

        # Capas GRU
        self.gru = nn.GRU(
            input_dim, hidden_dim, layer_dim, batch_first=True, dropout=dropout_prob
        )

        # Capa totalmente conectada
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        # Inicializando el estado oculto para la primera entrada con ceros
        h0 = torch.zeros(self.layer_dim, x.size(0), self.hidden_dim).requires_grad_()

        # Propagación hacia adelante pasando la entrada y el estado oculto al modelo.
        out, _ = self.gru(x, h0.detach())

        # Remodelar las salidas en forma de (batch_size, seq_length, hidden_size)
        # para que quepa en la capa totalmente conectada
        out = out[:, -1, :]

        # Convertir el estado final a nuestra forma de salida deseada (batch_size, output_dim)
        out = self.fc(out)

        return out
```

```python
def get_model(model, model_params):
    models = {
        "lstm": LSTMModel,
        "gru": GRUModel,
    }
    return models.get(model.lower())(**model_params)
```

# Optimizador

```python


device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"{device}" " is available.")

class Optimization:
    def __init__(self, model, loss_fn, optimizer):
        self.model = model
        self.loss_fn = loss_fn
        self.optimizer = optimizer
        self.train_losses = []
        self.val_losses = []

    def train_step(self, x, y):
        # Sets model to train mode
        self.model.train()

        # Makes predictions
        yhat = self.model(x)

        # Computes loss
        loss = self.loss_fn(y, yhat)

        # Computes gradients
        loss.backward()

        # Updates parameters and zeroes gradients
        self.optimizer.step()
        self.optimizer.zero_grad()

        # Returns the loss
        return loss.item()

    def train(self, train_loader, val_loader, batch_size=64, n_epochs=50, n_features=1):
        model_path = f'{self.model}_{datetime.now().strftime("%Y-%m-%d %H:%M:%S")}'

        for epoch in range(1, n_epochs + 1):
            batch_losses = []
            for x_batch, y_batch in train_loader:
                x_batch = x_batch.view([batch_size, -1, n_features]).to(device)
                y_batch = y_batch.to(device)
                loss = self.train_step(x_batch, y_batch)
                batch_losses.append(loss)
            training_loss = np.mean(batch_losses)
            self.train_losses.append(training_loss)

            with torch.no_grad():
                batch_val_losses = []
                for x_val, y_val in val_loader:
                    x_val = x_val.view([batch_size, -1, n_features]).to(device)
                    y_val = y_val.to(device)
                    self.model.eval()
                    yhat = self.model(x_val)
                    val_loss = self.loss_fn(y_val, yhat).item()
                    batch_val_losses.append(val_loss)
                validation_loss = np.mean(batch_val_losses)
                self.val_losses.append(validation_loss)

            if (epoch <= 10) | (epoch % 50 == 0):
                print(
                    f"[{epoch}/{n_epochs}] Training loss: {training_loss:.4f}\t Validation loss: {validation_loss:.4f}"
                )

        torch.save(self.model.state_dict(), model_path)

    def evaluate(self, test_loader, batch_size=1, n_features=1):
        with torch.no_grad():
            predictions = []
            values = []
            for x_test, y_test in test_loader:
                x_test = x_test.view([batch_size, -1, n_features]).to(device)
                y_test = y_test.to(device)
                self.model.eval()
                yhat = self.model(x_test)
                predictions.append(yhat.to(device).detach().numpy())
                values.append(y_test.to(device).detach().numpy())

        return predictions, values

    def plot_losses(self):
        plt.plot(self.train_losses, label="Training loss")
        plt.plot(self.val_losses, label="Validation loss")
        plt.legend()
        plt.title("Losses")
        plt.show()
        plt.close()


    def forecast_with_predictors(
        self, forecast_loader, batch_size=1, n_features=1, n_steps=100
    ):
        """Forecasts values for RNNs with predictors and one-dimensional output

        The method takes DataLoader for the test dataset, batch size for mini-batch testing,
        number of features and number of steps to predict as inputs. Then it generates the
        future values for RNNs with one-dimensional output for the given n_steps. It uses the
        values from the predictors columns (features) to forecast the future values.

        Args:
            forecast_loader (torch.utils.data.DataLoader): DataLoader that stores test data
            batch_size (int): Batch size for mini-batch training
            n_features (int): Number of feature columns
            n_steps (int): Number of steps to predict future values

        Returns:
            list[float]: The values predicted by the model

        """
        step = 0

        with torch.no_grad():
            predictions = []
            for x_test, _ in forecast_loader:
                x_test = x_test.view([batch_size, -1, n_features]).to(device)
                self.model.eval()
                yhat = self.model(x_test)
                predictions.append(yhat.to(device).detach().numpy())

                step += 1
                if step == n_steps:
                    break

        return predictions



    def forecast_with_lag_features(self, test_loader, batch_size=1, n_features=1, n_steps=100):
        test_loader_iter = iter(test_loader)
        predictions = []

        *_, (X, y) = test_loader_iter

        y = y.to(device).detach().numpy()
        X = X.view([batch_size, -1, n_features]).to(device)
        X = torch.roll(X, shifts=1, dims=2)
        X[..., -1, 0] = y.item(0)

        with torch.no_grad():
            self.model.eval()
            for _ in range(n_steps):
                X = X.view([batch_size, -1, n_features]).to(device)
                yhat = self.model(X)
                yhat = yhat.to(device).detach().numpy()
                X = torch.roll(X, shifts=1, dims=2)
                X[..., -1, 0] = yhat.item(0)
                predictions.append(yhat.item(0))

        return predictions
```
